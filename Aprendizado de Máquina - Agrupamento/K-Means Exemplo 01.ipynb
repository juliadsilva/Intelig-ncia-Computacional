#### Criar dados

```
from sklearn.datasets import make_blobs
```

```
data, _  = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=0.5, shuffle= True)
```

#### Análise Exploratória

```
import matplotlib.pyplot as plt
```

```
plt.rcParams["figure.dpi"] = 150
plt.scatter(data[:, 0], data[:,1], marker = "o", edgecolors= "black")
plt.show()
```

#### Criação do modelo K-Means


```
from sklearn.cluster import KMeans
```

```
modelo = KMeans(n_clusters=3)
modelo.fit(data)
```

#### Predições (Resposta do Modelo)

```
resposta = modelo.predict(data)
```

### Vizualizando a resposta


```
grupo_1 = data[resposta == 0]
grupo_2 = data[resposta == 1]
grupo_3 = data[resposta == 2]
centros = modelo.cluster_centers_
```

```
plt.scatter (grupo_1 [:,0], grupo_1 [:,1], marker = "o", c="blue", label = "Grupo 1", edgecolors= "black")
plt.scatter (grupo_2 [:,0], grupo_2 [:,1], marker = "s", c="green", label = "Grupo 2", edgecolors= "black")
plt.scatter (grupo_3 [:,0], grupo_3 [:,1], marker = "v", c="red", label = "Grupo 3", edgecolors= "black")
plt.scatter (centros [:,0], centros [:,1], marker = "x", c="pink", label = "Center")
plt.legend(scatterpoints=1)
plt.show()
```

### Metricas de Avaliação

```
from sklearn.metrics import calinski_harabasz_score, silhouette_score
sh = silhouette_score(data, modelo.labels_)
ch = calinski_harabasz_score(data,modelo.labels_)
```

```
print ("SH", sh)
print ("CH", ch)
```
